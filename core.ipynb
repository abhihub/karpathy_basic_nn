```python
def f(x):
    return 3*x**2 - 4*x +5
```

```python
xs = np.arange(-5, 5, 0.25)
ys = f(xs)
plt.plot(xs, ys)
```

```python
a = 2
b = -3
c = 10
d = a*b + c
print(d)

```

```python
h = 0.0001
d1 = a*b + c
c += h
d2 = a*b + c
slope = ((d2-d1)/h)

print ('d1',d1)
print ('d2', d2)
print ('slope', slope)
```
